```python
import pdfplumber
import re

nombre_pdf = list(uploaded.keys())[0]

texto_completo = ""
with pdfplumber.open(nombre_pdf) as pdf:
    for pagina in pdf.pages:
        texto_completo += pagina.extract_text() + "\n"

# Limpieza b√°sica
texto_limpio = re.sub(r'\[\d+\]', '', texto_completo)
texto_limpio = re.sub(r'\([A-Za-z, ]+\d{4}\)', '', texto_limpio)
texto_limpio = re.sub(r'[^a-zA-Z0-9√°√©√≠√≥√∫√±√Å√â√ç√ì√ö√ë\s\.,;:¬°!¬ø?\-]', '', texto_limpio)
texto_limpio = texto_limpio.lower()
texto_limpio = re.sub(r'\n\s*\n', '\n', texto_limpio)
texto_limpio = re.sub(r'\s+', ' ', texto_limpio).strip()
```

```python
from langdetect import detect
import ipywidgets as widgets
from IPython.display import display, clear_output
from transformers import MarianMTModel, MarianTokenizer, pipeline
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer
from docx import Document
from google.colab import files



idioma_detectado = detect(texto_limpio)
print(f"üåç El idioma detectado es: {idioma_detectado}")

opciones_traduccion = widgets.Dropdown(
    options=["Mantener idioma original", "Traducir a espa√±ol"],
    description="Acci√≥n:"
)

boton_generar = widgets.Button(description="üéØ Generar resumen ahora")

def al_clic(b):
    global texto_limpio, resumen_extractivo, resumen_abstractivo

    clear_output()
    print(f"üåç El idioma detectado es: {idioma_detectado}")
    if idioma_detectado == "en" and opciones_traduccion.value == "Traducir a espa√±ol":
        print("üîÅ Traduciendo al espa√±ol...")
        modelo_id = "Helsinki-NLP/opus-mt-en-es"
        tokenizer = MarianTokenizer.from_pretrained(modelo_id)
        model = MarianMTModel.from_pretrained(modelo_id)
        tokens = tokenizer(texto_limpio[:1000], return_tensors="pt", padding=True, truncation=True)
        traduccion = model.generate(**tokens)
        texto_limpio = tokenizer.decode(traduccion[0], skip_special_tokens=True)
        print("‚úÖ Traducci√≥n lista.")

    print("\nüìò Generando resumen extractivo...")
    parser = PlaintextParser.from_string(texto_limpio, Tokenizer("spanish"))
    resumidor = TextRankSummarizer()
    resumen_extractivo = "\n".join(str(oracion) for oracion in resumidor(parser.document, 5))
    print(resumen_extractivo)

    print("\nü§ñ Generando resumen abstractivo...")
    resumidor_ia = pipeline("summarization", model="facebook/bart-large-cnn")
    resumen_abstractivo = resumidor_ia(texto_limpio[:1024], max_length=130, min_length=30, do_sample=False)
    print(resumen_abstractivo[0]['summary_text'])

    # Guardar como .txt
    with open("resumen.txt", "w", encoding="utf-8") as f:
        f.write(resumen_extractivo)
    files.download("resumen.txt")

    # Guardar como .docx
    doc = Document()
    doc.add_heading("Resumen generado", level=1)
    doc.add_paragraph(resumen_extractivo)
    doc.save("resumen.docx")
    files.download("resumen.docx")

display(opciones_traduccion, boton_generar)
boton_generar.on_click(al_clic)
```
